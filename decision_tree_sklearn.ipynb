#**Membuat fungsi untuk melatih model, plotting model, dan mengetahui akurasi**

```
def get_result(model, X_train, X_test, y_train, y_test, X, y):
  model.fit(X_train, y_train)
  fig, ax = plt.subplots(figsize=(20,10))
  tree.plot_tree(model, feature_names=data.feature_names, class_names=data.target_names, ax=ax, filled=True)
  plt.show()
  return model.score(X_test, y_test)
```

#**Fungsi diaplikasikan untuk semua fold**

```
accur = []
i=1
dt = DecisionTreeClassifier(max_depth=4, min_samples_leaf=16, random_state=0, criterion='gini')
for train_index, test_index in Splitdata.split(X,Y):
  X_train, X_test, y_train, y_test = X[train_index], X[test_index], Y[train_index], Y[test_index]
  print('Fold ke : ',i)
  accur.append(get_result(dt, X_train,X_test,y_train,y_test,X,Y))
  print('Akurasi pada Fold ke {} = {} '.format(i, accur[i-1]))
  i+=1
```
